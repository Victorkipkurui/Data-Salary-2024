```python
sns.histplot(Data["salary_in_usd"], bins=30, kde=True)
plt.title("Distribution of Salary in USD")
plt.xlabel("Salary in USD")
plt.ylabel("Frequency")
plt.show()
```

```python
df = pd.DataFrame(Data)
modified_df = df.groupby(['experience_level', 'work_year'], as_index=False)['salary_in_usd'].mean()
plt.figure(figsize=(15, 6))
sns.barplot(x='work_year', y='salary_in_usd', hue='experience_level', data=modified_df)
plt.title('Bargraph Showing Average Salary by Experience per year')
plt.xlabel('Years')
plt.ylabel('Salary in USD')
plt.legend(title='Experience Level')
plt.show()
```

```python
df = pd.DataFrame(Data)
plt.figure(figsize=(15, 10))
colors = ['#FF5733', '#33FF57', '#3357FF', '#F4C724', '#24F4C7']
sns.boxplot(x='company_size', y='salary_in_usd', hue='work_year', data=df, palette =colors)
plt.title('Boxplot showing Distribution of Salaries by Company Size per year')
plt.xlabel('Company Size')
plt.ylabel('Salary in USD')
plt.legend(title='Work Year', bbox_to_anchor=(1, 1), loc='upper left')
plt.show()
```

```python
plt.figure(figsize=(10, 5))
colors = ['#FF5733', '#33FF57', '#24F4C7', '#FF33FF']
values = Data['employment_type'].value_counts()
sns.barplot(x='employment_type', y='salary_in_usd', hue="employment_type", data=Data, palette = colors)

for i, count in enumerate(values):
    plt.text(i, count + 10, f'{count}', ha='center', va='bottom')
     
plt.title('Salary by Employment Type')
plt.xlabel('Employment Type')
plt.ylabel('Salary in USD')
plt.show()
```

```python
modified_data1 = Data.groupby('job_title', as_index=False)['salary_in_usd'].mean()
modified_data1 = modified_data1.sort_values(by='salary_in_usd', ascending=False).head(30)
plt.figure(figsize=(15, 10))
sns.barplot(x='salary_in_usd', y='job_title', data=modified_data1)
plt.title('Average Salary by Job Title')
plt.xlabel('Salary in USD')
plt.ylabel('Job Title')
plt.show()
```

```python
modified_data = Data.groupby('company_location', as_index=False)['salary_in_usd'].mean()
modified_data = modified_data.sort_values(by='salary_in_usd', ascending=False).head(30)
plt.figure(figsize=(15, 10))
sns.barplot(x='salary_in_usd', y='company_location', data=modified_data)
plt.title('Average Salary by Company Location')
plt.xlabel('Salary in USD')
plt.ylabel('Company Location')
plt.show()
```
